### Species Distribution

```python
df["species"].value_counts()
```

### NumPy Statistics

```python
print("Mean Sepal Length:",
      np.mean(df["sepal_length"]))
```

```python
print("Median Petal Length:",
      np.median(df["petal_length"]))
```

```python
print("Standard Deviation:",
      np.std(df["sepal_length"]))
```

```python
print("Variance:",
      np.var(df["sepal_length"]))
```

### Correlation Analysis

```python
print(df.corr(numeric_only=True))
```

### Feature Engineering (Creating my own column)

```python
df["sepal_area"] = (
    df["sepal_length"] *
    df["sepal_width"]
)
print(df[["species","sepal_area"]].head())
```

### Group by Analysis

```python
species_summary = df.groupby("species").mean(numeric_only=True)

print(species_summary)
```

### Aggregate Functions

```python
df.groupby("species").agg(
    {
        "sepal_length":["mean","min","max"],
        "petal_length":["mean","min","max"],
        "petal_width":["mean","min","max"]
    }
)
```

### Sorting 

```python
largest_flowers=df.sort_values(
    by="petal_length",
    ascending=False
)
```

```python
largest_flowers.head()
```

### Filtering

```python
large_petals = df[
    df["petal_length"] > 5
]
```

```python
large_petals.head()
```

```python
species_summary.to_csv(
    "species_summary.csv"
)
```
